```python
import geopandas as gpd
import pandas as pd
import os, glob
```

```python
url = 'https://raw.github.com/ssujit/move_sustainable/main/data/'

# variable declaration
wkam = 'wk_am_hamburg.geojson' # wk_am: weekdays peak-traffic morning hours (6:00-8:59) 
wkpm = 'wk_pm_hamburg.geojson' # wk_pm: weekdays peak-traffic afternoon hours (14:00-16:59) 
satam = 'sat_am_hamburg.geojson' # satamo: Saturday peak-traffic morning hours (6:00-8:59) 
satpm = 'sat_pm_hamburg.geojson' # satpmo: Saturday peak-traffic afternoon hours (14:00-16:59)  
sunam = 'sun_am_hamburg.geojson' # sunamo: Saturday peak-traffic morning hours (6:00-8:59)
sunpm = 'sun_pm_hamburg.geojson' # sanpmo: Saturday peak-traffic afternoon hours (14:00-16:59) 

#read the files
wk_am = gpd.read_file(f'{url}{wkam}')
wk_pm = gpd.read_file(f'{url}{wkpm}')
sat_am = gpd.read_file(f'{url}{satam}')
sat_pm = gpd.read_file(f'{url}{satpm}')
sun_am = gpd.read_file(f'{url}{sunam}')
sun_pm = gpd.read_file(f'{url}{sunpm}')
```

```python
# List of geoDataFrames
geodataframes = [wk_am, wk_pm, sat_am, sat_pm, sun_am, sun_pm]

# Labels to each geoDataFrame
types = ['week', 'week', 'saturday', 'saturday', 'sunday', 'sunday']
times = ['AM', 'PM', 'AM', 'PM', 'AM', 'PM']
ct = ['Hamburg'] * len(geodataframes)

# Folder path to save the files
output_folder = 'dashboard_data'
if not os.path.exists(output_folder):
        os.makedirs(output_folder)

# Assigning labels to the 'Type' and 'Time' columns for each DataFrame
for gdf, week, duration, city in zip(geodataframes, types, times, ct):
    gdf['Type'] = week
    gdf['Time'] = duration
    gdf['City'] = city
    gdf['Type'] = gdf['Type'].astype(str)
    gdf['Time'] = gdf['Time'].astype(str)
    gdf['City'] = gdf['City'].astype(str)
    file_name = f"{week}_{duration}_hamburg.geojson"
    file_path = os.path.join(output_folder, file_name)
    gdf.to_file(file_path, driver='GeoJSON')
    #gdf.to_file(f"{week}_{duration}_hamburg.geojson", driver='GeoJSON')
```

```python
# Read GeoJSON files into GeoDataFrames for Berlin
ber_wk_am = gpd.read_file(url + 'wk_am_berlin.geojson')
ber_wk_pm = gpd.read_file(url + 'wk_pm_berlin.geojson')
ber_sat_am = gpd.read_file(url + 'sat_am_berlin.geojson')
ber_sat_pm = gpd.read_file(url + 'sat_pm_berlin.geojson')
ber_sun_am = gpd.read_file(url + 'sun_am_berlin.geojson')
ber_sun_pm = gpd.read_file(url + 'sun_pm_berlin.geojson')

# List of geoDataFrames for Berlin
ber_geodataframes = [ber_wk_am, ber_wk_pm, ber_sat_am, ber_sat_pm, ber_sun_am, ber_sun_pm]

# Labels for each geoDataFrame for Berlin
ber_types = ['week', 'week', 'saturday', 'saturday', 'sunday', 'sunday']
ber_times = ['AM', 'PM', 'AM', 'PM', 'AM', 'PM']
ber_ct = ['Berlin'] * len(ber_geodataframes)

# Folder path to save the files
output_folder = 'dashboard_data'

# Assigning labels and saving GeoDataFrames for Berlin
for ber_gdf, week, duration, city in zip(ber_geodataframes, ber_types, ber_times, ber_ct):
    ber_gdf['Type'] = week
    ber_gdf['Time'] = duration
    ber_gdf['City'] = city
    file_name = f"{week}_{duration}_berlin.geojson"
    file_path = os.path.join(output_folder, file_name)
    ber_gdf.to_file(file_path, driver='GeoJSON')
```

```python
ber_wk_am = gpd.read_file('week_AM_berlin.geojson')
ber_wk_pm = gpd.read_file('week_PM_berlin.geojson')
ber_sat_am = gpd.read_file('saturday_AM_berlin.geojson')
ber_sat_pm = gpd.read_file('saturday_PM_berlin.geojson')
ber_sun_am = gpd.read_file('sunday_AM_berlin.geojson')
ber_sun_pm = gpd.read_file('sunday_PM_berlin.geojson')

ham_wk_am = gpd.read_file('week_AM_hamburg.geojson')
ham_wk_pm = gpd.read_file('week_PM_hamburg.geojson')
ham_sat_am = gpd.read_file('saturday_AM_hamburg.geojson')
ham_sat_pm = gpd.read_file('saturday_PM_hamburg.geojson')
ham_sun_am = gpd.read_file('sunday_AM_hamburg.geojson')
ham_sun_pm = gpd.read_file('sunday_PM_hamburg.geojson')

```

```python
# Merge the two GeoDataFrames into a single GeoDataFrame
week_AM_merged_gdf = gpd.GeoDataFrame(pd.concat([ber_wk_am, ham_wk_am], ignore_index=True), crs=ber_wk_am.crs)
week_PM_merged_gdf = gpd.GeoDataFrame(pd.concat([ber_wk_pm, ham_wk_pm], ignore_index=True), crs=ber_wk_pm.crs)

sat_AM_merged_gdf = gpd.GeoDataFrame(pd.concat([ber_sat_am, ham_sat_am], ignore_index=True), crs=ber_sat_am.crs)
sat_PM_merged_gdf = gpd.GeoDataFrame(pd.concat([ber_sat_pm, ham_sat_pm], ignore_index=True), crs=ber_sat_pm.crs)

sun_AM_merged_gdf = gpd.GeoDataFrame(pd.concat([ber_sun_am, ham_sun_am], ignore_index=True), crs=ber_sun_am.crs)
sun_PM_merged_gdf = gpd.GeoDataFrame(pd.concat([ber_sun_pm, ham_sun_pm], ignore_index=True), crs=ber_sun_pm.crs)

# Save the merged GeoDataFrame to a new GeoJSON file
week_AM_merged_gdf.to_file("week_AM.geojson", driver="GeoJSON")
week_PM_merged_gdf.to_file("week_PM.geojson", driver="GeoJSON")
sat_AM_merged_gdf.to_file("sat_AM.geojson", driver="GeoJSON")
sat_PM_merged_gdf.to_file("sat_PM.geojson", driver="GeoJSON")
sun_AM_merged_gdf.to_file("sun_AM.geojson", driver="GeoJSON")
sun_PM_merged_gdf.to_file("sun_PM.geojson", driver="GeoJSON")
```
